```python
import sys
import pyodbc
class Restaurant():
    def __init__(self):
        try:
            self.conn=pyodbc.connect('Driver={SQL Server};' 'Server=MSTDPUNHPLT145;' 'Database=hotel;' 'Trusted_Connection=yes;')
            self.cursor=self.conn.cursor()
            self.cart={}
            print('database connectivity done successfully')
        except Exception as e:
            print('no connectivity',e)
    def show_menu(self):
        try:
            self.cursor.execute('select * from menu')
            print('this is menu')
            for i in self.cursor:
                print(i)
        except Exception as e:
            print('error for fetching menu',e)
    def take_order(self):
        while True:
            try:
                menu_id=int(input('enter menu id you want to buy'))
                qty=int(input('enter how many quanttites you want'))

                self.cursor.execute('select * from menu where menu_id=?',(menu_id,))
                item=self.cursor.fetchone()
                if item is None:
                    print('this menu is not in our menu card')
                    continue
                total=item[2]*qty
                # insert values into dictionary now
                self.cart[menu_id]={'name':item[1].strip(),'price':item[2],'qty':qty,'total':total}
                #insert values into orders
                self.cursor.execute('insert into orders  (menu_id,item_name,price,quantity,total) values(?,?,?,?,?)',(item[0],item[1],item[2],qty,total))
                self.conn.commit() # to save permannetly
            except:
                print(" System Exception Info:", sys.exc_info()[1])

            choice=input('do you want more order(Y/N): ').upper()
            if choice=='N':
                break
        return self.cart        
    def show_order_table(self):
        try:
            print('order table is as below')
            self.cursor.execute('select * from orders')
            for i in self.cursor :
                print(i)
        except:
            print('Exceptuon occured',sys.exc_info()[1])  
    
    def print_bill(self):

        try:
       

            # get latest order_id
            self.cursor.execute("select max(order_id) from orders")
            order_id = self.cursor.fetchone()[0]
            total=0
            print("\n" + "="*60)
            print("           WELCOME TO ABHIS HOTEL")
            print("="*60)
            print(f"Order ID : {order_id}")
            print("-"*60)
            print("MenuID  Item Name        Qty   Price   Total")
            print("-"*60)
            
            for menu_id, i in self.cart.items():
                print(f"{menu_id:<7} {i['name']:<15} {i['qty']:<5} {i['price']:<7} {i['total']}")
                total += i['total']
            print("-"*60)
            print(f"{'Grand Total':<35} {total}")
            print("="*60)
        except: 
            print(" System Exception Info:", sys.exc_info()[1])   
            
            
    def close_conn(self):
        try:
            self.conn.close()
            print('connection closed successfully')
        except:
            print('Exceptuon occured',sys.exc_info()[1])  
    


   


            
```
