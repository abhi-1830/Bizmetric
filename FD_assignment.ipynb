```python
import datetime
import sys

class Bank():

    def __init__(self, start_date, end_date):  # initialization of parameter
        try:
            print('welcome to ABHIS BANK')
            self.start = start_date
            self.end = end_date
            self.revised_date = datetime.datetime.strptime('09-05-2019', '%d-%m-%Y')
        except:
            print(sys.exc_info())

    def cal_day(self):
        try:
            print('how many days?')
            self.days = (self.end - self.start).days
            print('total days for fd done', self.days)
            return self.days
        except:
            print(sys.exc_info())

    def cal_interest(self, customer_type):
        try:
            self.customer_type = customer_type
            days = self.days

            if days < 7:
                print('no less fd duration plz')
                return

            if 7 <= days <= 45:
                bucket = '7 to 45 days'
                if customer_type == 'public':
                    intrest = 5.75
                else:
                    intrest = 6.25

            elif 46 <= days <= 179:
                bucket = '46 to 179 days'
                if customer_type == 'public':
                    intrest = 6.25
                else:
                    intrest = 6.75

            elif 180 <= days <= 210:
                bucket = '180 to 210 days'
                if customer_type == 'public':
                    intrest = 6.35
                else:
                    intrest = 6.85

            elif 211 <= days < 365:
                bucket = '211 days to less than 1 year'
                if customer_type == 'public':
                    intrest = 6.40
                else:
                    intrest = 6.90

            elif 365 <= days < 730:
                bucket = '1 year to less than 2 year'
                if customer_type == 'public':
                    if self.start < self.revised_date:
                        intrest = 6.80
                    else:
                        intrest = 7.00
                else:
                    if self.start < self.revised_date:
                        intrest = 7.30
                    else:
                        intrest = 7.50

            elif 730 <= days < 1095:
                bucket = "2 years to less than 3 years"
                if customer_type == "public":
                    if self.start < self.revised_date:
                        intrest = 6.80
                    else:
                        intrest = 6.75
                else:
                    if self.start < self.revised_date:
                        intrest = 7.30
                    else:
                        intrest = 7.25

            elif 1095 <= days < 1825:
                bucket = "3 years to less than 5 years"
                if customer_type == "public":
                    if self.start < self.revised_date:
                        intrest = 6.80
                    else:
                        intrest = 6.70
                else:
                    if self.start < self.revised_date:
                        intrest = 7.30
                    else:
                        intrest = 7.20

            elif 1825 <= days <= 3650:
                bucket = "5 years to 10 years"
                if customer_type == "public":
                    if self.start < self.revised_date:
                        intrest = 6.85
                    else:
                        intrest = 6.60
                else:
                    if self.start < self.revised_date:
                        intrest = 7.35
                    else:
                        intrest = 7.10
            else:
                print('input wrong')
                return

            return intrest, bucket

        except:
            print(sys.exc_info())
```
